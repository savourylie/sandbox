```python
from __future__ import division
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from IPython.display import display

df_main = pd.DataFrame({'A': range(100), 'B': np.random.rand(100)})
# display(df_main)

g = sns.PairGrid(df_main)
g.map_upper(plt.scatter, alpha=0.3)
g.map_lower(plt.scatter, alpha=0.3)
g.map_diag(sns.kdeplot, lw=3, legend=False)

plt.subplots_adjust(top=0.9)
g.fig.suptitle('THIS IS A TITLE, YOU BET') # can also get the figure from plt.gcf()
print(g)
# g.fig.axes[0].set_xlabel('12345')
# g.fig.axes[1].set_xlabel('67890')
```
